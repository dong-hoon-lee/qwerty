```python
import os
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

import warnings
warnings.filterwarnings('ignore')
```

```python
data_path = './data/'
rf_data_list = os.listdir(data_path+'rf_data/')

rf_list = []

for rf in rf_data_list:
    rf_list.append(rf.split('.')[0])
    
for rf in rf_list:
    globals()[f'data_{rf}'] = pd.read_csv(data_path+'rf_data/'+rf+'.csv')
```

```python
water_data_list = os.listdir(data_path+'water_data/')

water_list = []

for water in water_data_list:
    water_list.append(water.split('.')[0])
    
for water in water_list:
    globals()[f'water_{water}'] = pd.read_csv(data_path+'water_data/'+water+'.csv')
```
