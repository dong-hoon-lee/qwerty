```python
df_credits.columns = ['id', 'title', 'cast', 'crew']
df_credits_dropped = df_credits.drop(columns=['title'])
df_credits_dropped
```

```python
df_merged = df_movies.merge(df_credits_dropped, on='id')
df_merged
```

weighted vote rate
- weighted rate = **(v * R)/(v + m) + (m * C)/(v + m)**
    - v: number of votes
    - m: minimum number of votes(=threshold)
    - R: average rating of each movie
    - C: average rating of all movies

```python
C = df_merged['vote_average'].mean()
m = df_merged['vote_count'].quantile(0.75)
m
```

```python
q_movies = df_merged.copy().loc[df_merged['vote_count']>=m]
q_movies
```

```python
def weighted_rating(df, m=m, C=C):
    v = df['vote_count']
    R = df['vote_average']
    w_rating = (v/(v+m)*R)+(m/(v+m)*C)
    
    return w_rating
```

```python
q_movies['weighted_rating'] = q_movies.apply(weighted_rating, axis=1)
q_movies
```

```python
q_movies = q_movies.sort_values('weighted_rating', ascending=False)
q_movies
```
